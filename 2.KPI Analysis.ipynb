```python
import pandas as pd
import numpy as np
```

```python
# 1. 데이터 읽기
file_name = 'store_data.csv'
df = pd.read_csv(file_name, parse_dates=['주문 일자', '배송 일자']) # parse_dates -> 문자열 type을 바로 datetime type으로 변경
print(f"'{file_name}' 파일 로드 완료.\n")
```

```python
# 2. 총 고유 고객 수 계산
# '고객번호' 컬럼에서 중복을 제외한 고유한 값의 개수를 셉니다.
total_customers = df['고객번호'].nunique()
```

```python
# 3. 총 매출 계산
# '매출' 컬럼의 모든 값을 더합니다.
total_sales = df['매출'].sum()
```

```python
# 4. 총 수익 계산
# '수익' 컬럼의 모든 값을 더합니다.
total_profit = df['수익'].sum()
```

```python
# 5. 총 고유 주문 건수
total_orders = df['주문 번호'].nunique()
```

```python
# 6. 고객별 평균 지표 계산
# 6-1. 고객 1명당 평균 매출 (ARPU: Average Revenue Per User)
arpu = total_sales / total_customers

# 6-2. 고객 1명당 평균 수익
avg_profit_per_customer = total_profit / total_customers

# 6-3. 고객 1명당 평균 주문 횟수 (구매 빈도)
avg_orders_per_customer = total_orders / total_customers
```

```python
# 7. 결과 출력
print(f"--- '{file_name}' 파일 핵심 요약 ---")
print(f"총 고유 고객 수: {total_customers} 명")
print(f"총 매출: {total_sales:,.0f} 원") # 숫자를 1,000 단위 쉼표로 표시
print(f"총 수익: {total_profit:,.0f} 원") # 숫자를 1,000 단위 쉼표로 표시
print(f"총 고유 주문 건수: {total_orders} 건")
print(f"고객 1명당 평균 매출 (ARPU): {arpu:,.0f} 원")
print(f"고객 1명당 평균 수익: {avg_profit_per_customer:,.0f} 원")
print(f"고객 1명당 평균 주문 횟수: {avg_orders_per_customer:.2f} 회")
```
