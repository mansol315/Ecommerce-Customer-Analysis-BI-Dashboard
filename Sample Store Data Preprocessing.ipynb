## 전처리 PART

```python
import pandas as pd
import numpy as np
```

```python
# CSV 파일 읽기
superstore = '[KR]Superstore Sample.csv'
df = pd.read_csv(superstore)
```

```python
# 컬럼명 변경(특수문자 제거)
df = df.rename(columns={'반품?': '반품'})
```

```python
# 결측치 값 변경
df['반품'] = df['반품'].fillna('No')
```

```python
# 컬럼 type 변경 (문자열 type -> 날짜/시간 type)
df['주문 일자'] = pd.to_datetime(df['주문 일자'], format='%m/%d/%y')
df['배송 일자'] = pd.to_datetime(df['배송 일자'], format='%m/%d/%y')
```

```python
# 중복된 행 제거
duplicate_rows = df.duplicated().sum()
print(f"\n중복된 행의 수: {duplicate_rows}")

if duplicate_rows > 0:
    df = df.drop_duplicates()
    print(f"중복 제거 후 총 행의 수: {len(df)}")
```
